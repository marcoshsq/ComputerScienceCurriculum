## Processo de análise.

Agora que os dados estão limpos e padronizado, vamos começar a reponder as perguntas determinadas no ducumento de "Objetivos da análise". Começando pela primeira pergunta:

1. Como os membros anuais e os ciclistas casuais usam as bicicletas da Cyclistic de maneira diferente?
2. Por que os ciclistas casuais comprariam assinaturas anuais do Cyclistic?
3. Como a Cyclistic pode usar a mídia digital para influenciar os ciclistas casuais a se tornarem membros?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Converter colunas de data para o tipo datetime se ainda não estiverem
data_q1['start_time'] = pd.to_datetime(data_q1['start_time'])
data_q2['start_time'] = pd.to_datetime(data_q2['start_time'])
data_q3['start_time'] = pd.to_datetime(data_q3['start_time'])
data_q4['start_time'] = pd.to_datetime(data_q4['start_time'])

# Criar coluna 'hour' para representar a hora do dia
data_q1['hour'] = data_q1['start_time'].dt.hour
data_q2['hour'] = data_q2['start_time'].dt.hour
data_q3['hour'] = data_q3['start_time'].dt.hour
data_q4['hour'] = data_q4['start_time'].dt.hour

# Configurações visuais
sns.set_style("whitegrid")
sns.set_palette("viridis")

# Plotar padrões de uso ao longo do dia
plt.figure(figsize=(14, 16))

plt.subplot(4, 1, 1)
sns.countplot(x='hour', data=data_q1, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q1')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.subplot(4, 1, 2)
sns.countplot(x='hour', data=data_q2, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q2')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.subplot(4, 1, 3)
sns.countplot(x='hour', data=data_q3, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q3')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.subplot(4, 1, 4)
sns.countplot(x='hour', data=data_q4, hue='usertype')
plt.title('Padrões de Uso ao Longo do Dia - Q4')
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.legend(title='Tipo de Usuário')

plt.tight_layout()
plt.savefig('padroes_de_uso_aprimorado.png')
plt.show()

# Seu código para preparar e plotar os padrões de uso ao longo do dia...

plt.savefig('padroes_de_uso.png')
plt.show()
```

# Análise dos Padrões de Uso de Membros Anuais e Ciclistas Casuais

## Q1 (Primeiro Trimestre):
- Ambos os grupos têm um aumento de uso pela manhã, indicando prováveis deslocamentos para o trabalho ou outras atividades diurnas.
- Ciclistas casuais mostram uma tendência a picos de uso durante a tarde e no início da noite, indicando possível uso recreativo.
- Membros anuais mantêm uma distribuição mais estável ao longo do dia.

## Q2 (Segundo Trimestre):
- O padrão geral é semelhante ao Q1, mas com uma tendência mais clara de uso recreativo para ciclistas casuais durante a tarde.
- Membros anuais continuam com uma distribuição relativamente estável.

## Q3 (Terceiro Trimestre):
- Tanto para membros anuais quanto para ciclistas casuais, há uma tendência mais pronunciada de aumento de uso durante a tarde, indicando possível uso recreativo ou turístico.
- Ciclistas casuais mostram um pico significativo durante a tarde.

## Q4 (Quarto Trimestre):
- Os padrões são semelhantes aos trimestres anteriores, mas com uma queda geral no uso no final do ano, especialmente para ciclistas casuais.

**Observações Gerais:**
- Ciclistas casuais tendem a mostrar variações mais acentuadas durante o dia, sugerindo uso mais flexível e recreativo.
- Membros anuais mantêm padrões mais estáveis, indicando possíveis deslocamentos regulares.

Essa análise inicial fornece insights sobre como diferentes grupos usam as bicicletas ao longo do dia em diferentes trimestres do ano.

### Padrões Semanais de Uso das Bicicletas Cyclistic

#### Q1 (Primeiro Trimestre):
- Ambos os grupos (membros anuais e ciclistas casuais) mostram uma diminuição relativa no uso das bicicletas aos domingos e segundas-feiras.

- Membros anuais mantêm uma distribuição mais estável ao longo da semana, indicando possíveis deslocamentos regulares.

#### Q2 (Segundo Trimestre):
- Padrões semanais semelhantes ao Q1, mas com uma tendência mais acentuada de aumento de uso para ciclistas casuais durante a semana, atingindo o pico na quarta-feira.
- Membros anuais continuam a mostrar uma distribuição relativamente estável.

#### Q3 (Terceiro Trimestre):
- Ambos os grupos mantêm a tendência de aumento de uso durante a semana, mas com uma diferença mais evidente entre ciclistas casuais e membros anuais.
- Ciclistas casuais têm um pico notável na quinta-feira, enquanto membros anuais mantêm uma distribuição estável.

#### Q4 (Quarto Trimestre):
- Padrões semanais semelhantes aos trimestres anteriores, mas com uma queda geral no uso no final do ano, especialmente aos domingos e segundas-feiras.
- Ciclistas casuais ainda mostram um pico notável nas quintas-feiras.

**Observações Gerais:**
- Os padrões semanais sugerem que ciclistas casuais podem usar as bicicletas Cyclistic mais frequentemente durante a semana, possivelmente para atividades recreativas ou deslocamentos diários.
- Membros anuais parecem manter uma frequência mais constante ao longo da semana, indicando um uso mais regular e consistente.
